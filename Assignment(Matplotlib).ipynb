```python
fig,axes=plt.subplots(nrows=1,ncols=2)
plt.subplot(1,2,1)
plt.plot(x,y,'b--',linewidth=3)
plt.subplot(1,2,2)
plt.plot(x,z,'r',linewidth=3)
```
